```python
import csv
import re

#Define the variable
paragraph_list = []
word_list = []
sentence_list = []
word_dictionary = {}
sentence_dictionary = {}

letter_count = 0
word_count = 0
word_in_sentence_count = 0
average_letter_Count = 0
average_word_in_sentence_count = 0

#Define the text file location
file = "raw_data/paragraph_1.txt"

#read the text file and covert it into a list
with open(file,'r') as text:
    paragraph = text.read()
    paragraph_list = list(paragraph)
#identify the last character in the paragraph
    last_char = (paragraph_list[-1])

#obtain the word count in the paragraph
word_count = len(paragraph.split())

#obtain the sentence count in the paragraph. 
#Delimiter is used as period to identify the nbr of sentences(.)
#if the last character is "." it will return additional one length and needs to be subtracted from the total length

if last_char == '.':
    sentence_count = len(paragraph.split(".")) - 1
else:
    sentence_count = len(paragraph.split("."))

#create a list of words in the paragraph
word_list = list(paragraph.split())

#For the word_list create a dictionary consisting of the word and its length. 
#This will give the letter count per word

for word in word_list:
    if word in word_dictionary:
        word_dictionary[word] = word_dictionary[word] + len(word)
    else:
        word_dictionary[word] = len(word)

#Scan through the dictionary Obtain the total letter count. 
for key,value in word_dictionary.items():
    letter_count = letter_count + int(value)

#Determine the avergae average_letter count
average_letter_count = letter_count / word_count

#create a list of sentences in the paragraph
sentence_list = list(paragraph.split("."))

#For the sentence_list create a dictionary consisting of the sentences and length of nbr of words. 
#This will give the letter count per word

for word in sentence_list:
    if word in sentence_dictionary:
        sentence_dictionary[word] = sentence_dictionary[word] + len(word.split())
    else:
        sentence_dictionary[word] = len(word.split())

#Scan through the dictionary Obtain the total words per sentence . 
for key,value in sentence_dictionary.items():
    word_in_sentence_count = word_in_sentence_count + int(value)

#Determine the avergae average_word_in_sentence count
average_word_in_sentence_count  = word_in_sentence_count /  sentence_count

print("----------------------------------")
print("Paragraph Analysis")
print("----------------------------------")
print("Approximate Word Count: " + str(word_count))
print("Approximate Sentence Count: " + str (sentence_count))
print ("Average Letter Count: " + str(average_letter_count))
print ("Average Sentence Length: " + str(average_word_in_sentence_count))

#Define the output file loction
outfile = "raw_data/output.txt"

with open (outfile,'w') as text:
    text.write("Paragraph Analysis")
    text.write("\n----------------------------------")
    text.write("\nApproximate Word Count: " + str(word_count))
    text.write("\nApproximate Sentence Count: " + str (sentence_count))
    text.write("\nAverage Letter Count: " + str(average_letter_count))
    text.write("\nAverage Sentence Length: " + str(average_word_in_sentence_count))
```
